Self Implementation

```python
# Opening the file
file =  open(filename, "r")

# Storing all lines as an array
lines = file.readlines()

# Close file
file.close()
```

```python
# Reading meta data
meta_i, meta_j = 10, 16
meta_lines = lines[meta_i: meta_j]
meta_data = {}

for line in meta_lines:
    index = line.find("!")
    val, key = line[:index], line[index + 1:]
    flag = False
#     Removing excessive whitespace characters in val
    j = index
    for i in range(1, len(val)):
        if val[i] == val[i - 1] and val[i] == ' ':
            j = i - 1
            break
    val = val[:j]
#     Removing \n characters in key
    key = key[:-1]
#     Adding value in metadata dictionary
    meta_data[key] = val

meta_data
```

```python
cols = lines[7].split()
```

```python
cols.insert(0, 'Configuration')
```

```python
cols
```

```python
cols.remove('Hz')
cols[3] += ' Hz'
cols
```

```python
# Reading Energy Levels
energy_i = 17
energy_j = energy_i + int(meta_data['Number of energy levels'])
energy_lines = lines[energy_i: energy_j]
energy_data = []

index = 0
for line in energy_lines:
    vals = line.split()
#     Converting strings to floats
    for i in range(1, len(vals)):
        vals[i] = float(vals[i])
    energy_data.append(vals)

energy_data = pd.DataFrame(data = energy_data, columns = cols)
```

```python
cols2 = lines[184].split()
cols2
```

```python
# Rectifying Columns
cols2.remove('%')
cols2[6] = '% ' + cols2[6]
cols2.insert(5, 'j')
cols2[4] = 'i'
cols2[0] = 'State A'
cols2.insert(1, 'State B')
```

```python
cols2
```

```python
# Reading Oscillator Strengths
osc_i = 186
osc_j = osc_i + int(meta_data['Number of transitions'])
osc_lines = lines[osc_i: osc_j]
osc_data = []

index = 0
for line in osc_lines:
    vals = line.split()
#     To check if the two states are not combined
    flag = vals[0].find(']') + 1 == len(vals[0])
    if not flag:
        vals.insert(1, vals[0][vals[0].find(']') + 1:])
        vals[0] = vals[0][:vals[0].find(']') + 1]
        vals = vals[:-1]
    vals[1] = vals[1][1:]
    vals[5] = vals[5][:-1]
#     Converting strings to floats
    for i in range(2, len(vals)):
        try:
            vals[i] = float(vals[i])
        except ValueError:
            vals[i] =  np.nan
    vals = vals[:-1]
    osc_data.append(vals)

osc_data = pd.DataFrame(data = osc_data, columns = cols2)
```
